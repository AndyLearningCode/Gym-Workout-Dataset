```r
# Library Imports
library(tidyverse)
library(corrplot)
library(ggplot2)
library(car)
#if needed
#install.packages("corrplot", dependencies=TRUE, repos='http://cran.rstudio.com/')
```

```r
df <- read_csv("gym_members_exercise_tracking.csv")
summary(df)
```

```r
# check for NA
colSums(is.na(df))
```

```r
df <- df %>%
  mutate(Gender = as.factor(Gender),
         Workout_Type = as.factor(Workout_Type),
         Experience_Level = as.factor(Experience_Level))
```

```r
options(repr.plot.width = 6, repr.plot.height = 5)
ggplot(df, aes(x = Calories_Burned)) +
  geom_histogram(binwidth = 50, fill = "skyblue", color = "black") +
  labs(title = "Calories Burned")

ggplot(df, aes(x = Workout_Type, y = Calories_Burned, fill = Workout_Type)) +
  geom_boxplot() +
  labs(title = "Calories Burned by Workout Type")

ggplot(df, aes(x = `Session_Duration (hours)`, y = Calories_Burned)) +
  geom_point() +
  geom_smooth() +
  labs(title = "Session Duration vs Calories Burned")
```

# Base model + motivation for feature selection

```r
base_model <- lm(Calories_Burned ~ ., data = df)
summary(base_model)
```

```r
par(mfrow = c(2, 2))
plot(base_model)
par(mfrow = c(1,1))
```

```r
vif(base_model)
```

```r
# Correlation matrix for numeric variables
options(repr.plot.width = 6, repr.plot.height = 5)
num_vars <- df %>% 
    select(where(is.numeric)) %>% 
    select(-Calories_Burned)
corrplot(cor(num_vars), method = "color", type = "upper")
```
